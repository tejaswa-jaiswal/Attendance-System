```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/SMSSpamCollection.txt', sep = '\t', names= ['label', 'messages'])
```

```python
print(len(df))
df
y= df['label']
```

```python
import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
stopwords.words('english')
```

```python
corpus = []
for i in range(0,  len(df)):
  new = re.sub('[^a-zA-Z0-9]', ' ', df['messages'][i])
  new = new.lower()
  new = new.split()

  new = [ps.stem(word) for word in new if word not in stopwords.words('english')]
  new = ' '.join(new)
  corpus.append(new)

```

```python
corpus
```

**firstly doing with bag of words**

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=2500, binary= True, ngram_range=(2,2))
X = cv.fit_transform(corpus).toarray()
```

```python
X.shape
```

```python
X
```

```python
y = pd.get_dummies(df['label'])
y = y.iloc[:,1].astype(int)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train ,  y_test = train_test_split(X, y , test_size= 0.2)
```

```python
from sklearn.naive_bayes import MultinomialNB
nb = MultinomialNB()
nb.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report
y_pred = nb.predict(x_test)
acc = accuracy_score(y_test , y_pred)
cl_rpt = classification_report(y_test, y_pred)
print(acc)
print(" ----------------- ")
print(cl_rpt)
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
nltk.download('wordnet')
```

```python
corpus = []
for i in range(0, len(df)):
  review = re.sub('[^a-zA-Z0-9]', ' ', df['messages'][i])
  review = review.lower()
  review = review.split()
  review = [lemmatizer.lemmatize(word) for word in review if not word in stopwords.words('english')]
  review = ' '.join(review)
  corpus.append(review)
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
```

```python
[simple_preprocess(word) for word in sent_tokenize('He is a boy. He is good at basketball')]
```

```python
len(corpus)
```

```python
corpus
```

```python
words = []
for sent in corpus:
  tokens = sent_tokenize(sent)
  for sent in tokens:
    words.append(simple_preprocess(sent))
```

```python
len(words)
```

```python
import gensim
model = gensim.models.Word2Vec(words, window= 5, min_count=2)

```

```python
def avg_word2vec(doc):
    # remove out-of-vocabulary words
    #sent = [word for word in doc if word in model.wv.index_to_key]
    #print(sent)

    return np.mean([model.wv[word] for word in doc if word in model.wv.index_to_key],axis=0)
                #or [np.zeros(len(model.wv.index_to_key))], axis=0)
```

```python

from tqdm import tqdm
```

```python
vec = []
for message in tqdm(corpus):
    # Process each message individually
    processed_message = simple_preprocess(message)
    avg_vector = avg_word2vec(processed_message)
    vec.append(avg_vector)

# Convert vec to a numpy array
vec

# y already has the correct number of samples (5572) and is aligned with the original corpus
# So, no need to modify y here.
```
